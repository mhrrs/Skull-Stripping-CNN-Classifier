```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv = np.loadtxt("./image_data/BSE_BET_DATA/labels.csv", delimiter=",", dtype=str)

yy = 0
yn = 0
ny = 0
nn = 0

for item in csv:
    
    if item[1] == 'Yes' and item[2] == 'Yes':
        yy = yy + 1
    
    if item[1] == 'Yes' and item[2] == 'No':
        yn = yn + 1
        
    if item[1] == 'No' and item[2] == 'Yes':
        ny = ny + 1
        
    if item[1] == 'No' and item[2] == 'No':
        nn = nn + 1
    
```

```python
print("yy: ", yy, "\nyn: ", yn, "\nny: ", ny, "\nnn: ", nn)
```

```python
data = {csv[0][1]: yn, csv[0][2]: ny, "Perfect": nn}
```

```python
x = data.keys()
y = list(data.values())


#fig, = plt.subplots(figsize = (10,5))
plt.bar(x, [0,0, 125+15], color='pink', width=0.4, label="IXI Combined with NFBS")
plt.bar(x, y, color='orchid', width=0.4, label="IXI Dataset Only")


plt.xlabel('Scan Label')
plt.ylabel('Number of Samples')
plt.title('BSE_BET_IXI Dataset Summary')

for i in range(len(x)):
    plt.text(i, y[i]+ 10, y[i], color="gray")

plt.text(2, 150, 125+15, color="gray")
    
plt.legend(loc=5)
    
plt.show()
```
